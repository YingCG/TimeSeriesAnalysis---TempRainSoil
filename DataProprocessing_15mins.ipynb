```python
import pandas as pd
```

# Load the datasets

```python
air_temperature = pd.read_csv("../AucklandCouncilOpenData_08June24/AirTempHourlyAggregate-cleaned.csv")
air_temperature
```

```python
rainfall = pd.read_csv("../AucklandCouncilOpenData_08June24/Rainfall-cleaned.csv")
rainfall
```

```python
soil_moisture = pd.read_csv("../AucklandCouncilOpenData_08June24/SoilMoisture-cleaned.csv")
soil_moisture
```

```python
soil_temperature = pd.read_csv('../AucklandCouncilOpenData_08June24/SoilTemp-Cleaned.csv')
soil_temperature
```

# Convert to Datetime

```python
air_temperature['Timestamp (UTC+12:00)'] = pd.to_datetime(air_temperature['Timestamp (UTC+12:00)'], format='%m/%d/%Y %H:%M')
rainfall['Timestamp (UTC+12:00)'] = pd.to_datetime(rainfall['Timestamp (UTC+12:00)'], format='%m/%d/%Y %H:%M')
soil_moisture['Timestamp (UTC+12:00)'] = pd.to_datetime(soil_moisture['Timestamp (UTC+12:00)'], format='%m/%d/%Y %H:%M')
soil_temperature['Timestamp (UTC+12:00)'] = pd.to_datetime(soil_temperature['Timestamp (UTC+12:00)'], format='%m/%d/%Y %H:%M')
```

```python
air_temperature.info()
rainfall.info()
soil_moisture.info()
soil_temperature.info()
```

```python
# Set timestamp as the index
for df in [air_temperature, rainfall, soil_moisture, soil_temperature]:
    df.set_index('Timestamp (UTC+12:00)', inplace=True)
```

# Resampling data to same frequency - 15-minute intervals

```python
print(air_temperature.index[air_temperature.index.duplicated()])
print(rainfall.index[rainfall.index.duplicated()])
print(soil_moisture.index[soil_moisture.index.duplicated()])
print(soil_temperature.index[soil_temperature.index.duplicated()])# Drop duplicates
air_temperature = air_temperature[~air_temperature.index.duplicated(keep='first')]
rainfall = rainfall[~rainfall.index.duplicated(keep='first')]
soil_moisture = soil_moisture[~soil_moisture.index.duplicated(keep='first')]
soil_temperature = soil_temperature[~soil_temperature.index.duplicated(keep='first')]

air_temperature = air_temperature.resample('15T').ffill()
rainfall = rainfall.resample('15T').ffill()
soil_moisture = soil_moisture.resample('15T').mean().interpolate(method='linear')
soil_temperature = soil_temperature.resample('15T').interpolate(method='linear')
```

# linear interpolation for consistency

```python
air_temperature.interpolate(method='linear', inplace=True)
rainfall.interpolate(method='linear', inplace=True)
soil_moisture.interpolate(method='linear', inplace=True)
soil_temperature.interpolate(method='linear', inplace=True)
```

# Merge datasets

```python
temp_rain_soil = air_temperature.join([rainfall, soil_moisture, soil_temperature], how='inner')
temp_rain_soil
```

# Drop Missing Values

```python
temp_rain_soil.dropna(inplace=True)
temp_rain_soil.info()
```
